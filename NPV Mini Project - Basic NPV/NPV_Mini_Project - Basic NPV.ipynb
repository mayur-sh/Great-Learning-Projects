```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly as px
```

```python
df1 = pd.DataFrame([[1,"1979 [1978]","K. K.","Mickéy Mousé"],
        [2,"1868","L., L. L.","Donald Duck"],
        [218,"1869","RemaÊ¿.","Mini Mouse"],
        [472,"1851","A'BECKETT, Gilbert Abbott.","Donald Duck"],
        [5,"1857","ABATI, Francesco.","Pink Panther"]],columns=["Identifier","Date","Code_For_Name","Full_Name"])
df1
```

####  1.A) Clean the `date` column as shown in below format without using replace function.

```python
df1a = df1.Date.str.extract(r"^(\d{4})",expand=False)

df1.Date=pd.to_numeric(df1a)
df1
```

###### 1.B) Clean the `Code_For_Name` column as shown in below format without using replace function.

```python
for v,i in zip(df1.index,df1.Code_For_Name):
    i = "".join(x for x in i if ord(x)>= 65 and ord(x) <=90  or  ord(x)>= 97 and ord(x) <=122  )
    df1.Code_For_Name[v] = i

df1
```

###### 1.C) Clean the `Full_Name` column as shown in below format.


```python

df1.Full_Name.str.split()
df1["First_Name"] = ""
df1["Last_Name"] = ""

for v,i in zip(df1.index,df1.Full_Name.str.split()):
    df1.First_Name[v] = i[0]
    df1.Last_Name[v]  = i[1]


for v,i in zip(df1.index,df1.First_Name):
    i = "".join(x for x in i if ord(x)>= 65 and ord(x) <=90  or  ord(x)>= 97 and ord(x) <=122  )
    df1.First_Name[v] = i
for v,i in zip(df1.index,df1.Last_Name):
    i = "".join(x for x in i if ord(x)>= 65 and ord(x) <=90  or  ord(x)>= 97 and ord(x) <=122  )
    df1.Last_Name[v] = i

df1.drop("Full_Name",axis=1,inplace=True)
df1
```

#### F. If you order grilled chicken rather than crispy chicken, what is the change in calorie intake?
        Step1:  Display the suitable graph for the number of calories in Crispy Chicken category

        Step2: Find the average calories for Crispy Chicken Category

        Step3: Display the suitable graph for the number of calories in Grilled Chicken category

        Step4: Find the average calories for Grilled Chicken Category

        Step5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

        Step6: Write your observations.

```python
ccl = ml.filter([v for v,i in zip(ml.Item_in_Menu.index,ml.Item_in_Menu.values) if "Crispy Chicken" in i],axis=0) 
cgl = ml.filter([v for v,i in zip(ml.Item_in_Menu.index,ml.Item_in_Menu.values) if "Grilled Chicken" in i],axis=0)
```

##### Step1: Display the suitable graph for the number of calories in Crispy Chicken category

```python
plt.bar(x=ccl.Item_in_Menu,height=ccl.Calories)
plt.xticks(rotation=90)
plt.show()
```

##### Step2: Find the average calories for Crispy Chicken Category

```python
print("The average calories for Crispy Chicken Category is ",round(ccl.Calories.mean(),2))
```

##### Step3: Display the suitable graph for the number of calories in Grilled Chicken category

```python
plt.bar(x=cgl.Item_in_Menu,height=cgl.Calories)
plt.xticks(rotation=90)
plt.show()
```

##### Step4:Find the average calories for Grilled Chicken Category

```python
print("The average calories for Grilled Chicken Category is ",round(cgl.Calories.mean(),2))
```

##### Step 5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

```python
plt.figure(figsize=(10,8),dpi=100)
plt.bar(x=ccl.Item_in_Menu,height=ccl.Calories,label="Crispy Chicken")
plt.bar(x=cgl.Item_in_Menu,height=cgl.Calories,label="Grilled Chicken")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

##### Step6: Write your observations

```python
ctc=(pd.DataFrame(ml.groupby(by="Category_Type")["Calories"].sum()))

ctc.sort_values(by="Calories",ascending=False,inplace=True)

ctc
```

```python
plt.bar(x=ctc.index,height=ctc.Calories)
plt.xticks(rotation=90)
plt.show()

print(f"As per our observation, the consumption of Snacks & Sides result in most calories intake ie. {ctc.Calories.max()} \nAnd the lowest Calorie intake ie. {ctc.Calories.min()} is by consumpttion of Breakfast. ")
```
